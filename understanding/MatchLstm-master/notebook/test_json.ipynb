```python
import os
import json
```

```python
def read_json_file(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            line = str.lower(line)
            line = json.loads(line)
            data.append(line)
    return data
```

```python
train_data = read_json_file(os.path.join('..', 'dataset', 'snli_1.0_train.jsonl'))
```

```python
for d in train_data:
    if '' in d['annotator_labels']:
        print(d)
```
